```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-9b-it")

model = AutoModelForCausalLM.from_pretrained("google/gemma-2-9b-it", quantization_config=quantization_config)
```

```python
#entra de texto
input_text = "indicame las caracteristicas de python en formato de lista ordenado del mayor beneficio al menor"

#return_tensors "pt" para convertir en pytorch que es lo que acepta modelo
#cuda mueve el tensor a la GPU
input_ds = tokenizer(input_text, return_tensors="pt").to("cuda")

#limitar la generacion del toke
output = model.generate(**input_ds, max_new_tokens=1024)
#decodifica los tokens generados a un medio legible
print(tokenizer.decode(output[0]))
```
