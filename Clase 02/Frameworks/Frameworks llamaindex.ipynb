Ejemplo unificado

```python
models = [
    {"class": OpenAI, "model": "gpt-4", "api_key": apikey_openai},
    {"class": Anthropic, "model": "claude-opus-4-20250514", "api_key": apikey_claude},
    {"class": Gemini, "model": "models/gemini-2.5-flash", "api_key": apikey_gemini},
]

# Pregunta que deseas enviar
question = "¿Cuántos años tiene el mundo?"

# Iterar sobre los modelos
for model_config in models:
    llm = model_config["class"](temperature=0, model=model_config["model"], api_key=model_config["api_key"])
    chat_engine = SimpleChatEngine.from_defaults(llm=llm)
    response = chat_engine.chat(question)
    print("--------")
    print(f"Modelo: {model_config['model']}")
    print(response.response)  # Muestra la respuesta de cada modelo
```
